```python
import random

# Function to generate a random array of size X * Y
def generate_random_costs(x, y, min_value=1, max_value=150, seed=None):
    if seed is not None:
        random.seed(seed)  # Set seed for reproducibility
    return [[random.randint(min_value, max_value) for _ in range(y)] for _ in range(x)]

# Define the dimensions (X * Y)
X = 45  # For example, 5 rows
Y = 30  # For example, 4 columns

# Specify the seed
seed_value = 42

# Generate random costs array with the seed
costs = generate_random_costs(X, Y, seed=seed_value)

# Format and write to file
output_file = f"formatted_costs_with_seed_{X}_{Y}.txt"

with open(output_file, "w") as file:
    file.write("costs = [\n")
    for row in costs:
        file.write(f"    {row},\n")
    file.write("]")

print(f"Random costs array with seed {seed_value} formatted and saved to {output_file}")
```
